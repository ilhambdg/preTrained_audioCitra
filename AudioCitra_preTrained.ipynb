```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe_audio = pipeline("audio-classification", model="Hemg/violence-audio-Recognition-1111")

pipe_video = pipeline("video-classification", model="MCG-NJU/videomae-base-short-finetuned-ssv2")
```

```python
hasil_audio = pipe_audio("/masukan_path_fileAudio")
print(hasil_audio)

hasil_video = pipe_video("/masukan_path_fileVideo")
print(hasil_video)
```
